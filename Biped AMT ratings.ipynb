# Use rankings instead of specific numerical ratings

Infer rankings from specific numerical values. Each set has 5 ratings. Use those ratings to instead construct rankings.

More-or-less works out the same as going purely based on averages of the numerical ratings.

```python
from collections import defaultdict

biped_rankings = defaultdict(list)
biped_avg_rankings = defaultdict(float)

for i, group in biped_ratings.groupby("amt_rating_set"):
    for rank, (j, row) in enumerate(group.sort("amt_user_rating", ascending=False).iterrows()):
        biped_rankings[row["biped_num"]].append(rank + 1)
        
for biped in biped_rankings.keys():
    biped_avg_rankings[biped] = mean(biped_rankings[biped])

# bootstrap.ci is acting weird when 22 has all 1's
biped_rankings[22][0] = 1.000001
```

```python
import scikits.bootstrap as bootstrap
import scipy

biped_nums = []

figure(figsize=(20, 7))

for i, biped in enumerate(sorted(biped_avg_rankings, key=biped_avg_rankings.get)):
    biped_nums.append(biped)
    conf_int = bootstrap.ci(data=biped_rankings[biped], statfunction=scipy.mean)
    errorbar(x=[i], y=[biped_avg_rankings[biped]],
             yerr=[[biped_avg_rankings[biped] - conf_int[0]],
                   [conf_int[1] - biped_avg_rankings[biped]]],
             fmt="o")
    
xticks(range(len(biped_nums)), biped_nums, fontsize=14)
yticks(fontsize=14)
xlabel("Biped number", fontsize=16)
ylabel("Average ranking", fontsize=16)
xlim(-1, 49)
ylim(0.5, 4)
```

# Alternative ranking systems

Averaging the rankings loses a lot of information.

Try instant-runoff voting system: https://en.wikipedia.org/wiki/Instant-runoff_voting

Try creating global ranking list that minimizes sum of Kendall tau distance to all other user ranking lists: https://en.wikipedia.org/wiki/Kendall_tau_distance

## IRV ranking system - bottom rank first

```python
from collections import defaultdict

biped_rankings = []

for i, group in biped_ratings.groupby("amt_rating_set"):
    rankings = []
    for rank, (j, row) in enumerate(group.sort("amt_user_rating", ascending=False).iterrows()):
        rankings.append(row.biped_num)
    
    if len(rankings) > 0:
        biped_rankings.append(rankings)


global_ranking_list = []

while len(biped_rankings) > 0:
    votes = defaultdict(int)
    
    for rankings in biped_rankings:
        votes[rankings[0]] += 1
        
    bottom_voted = sorted(votes, key=votes.get, reverse=False)[0]
    
    global_ranking_list.insert(0, bottom_voted)
    
    new_biped_rankings = []
    for rankings in biped_rankings:
        rankings = [x for x in rankings if x != bottom_voted]
        if len(rankings) > 0:
            new_biped_rankings.append(rankings)
            
    biped_rankings = new_biped_rankings
```

## IRV ranking system - top rank first

```python
from collections import defaultdict

biped_rankings = []

for i, group in biped_ratings.groupby("amt_rating_set"):
    rankings = []
    for rank, (j, row) in enumerate(group.sort("amt_user_rating", ascending=False).iterrows()):
        if row.biped_num not in rankings:
            rankings.append(row.biped_num)
    
    if len(rankings) > 0:
        biped_rankings.append(rankings)


global_ranking_list = []

while len(biped_rankings) > 0:
    votes = defaultdict(int)
    
    for rankings in biped_rankings:
        votes[rankings[0]] += 1
        
    top_voted = sorted(votes, key=votes.get, reverse=True)[0]
    
    global_ranking_list.append(top_voted)
    
    new_biped_rankings = []
    for rankings in biped_rankings:
        rankings = [x for x in rankings if x != top_voted]
        if len(rankings) > 0:
            new_biped_rankings.append(rankings)
            
    biped_rankings = new_biped_rankings
```

## Hillclimber optimization

```python
from scipy.stats import kendalltau
import random

all_biped_nums = set()
biped_rankings = []

for i, group in biped_ratings.groupby("amt_rating_set"):
    rankings = []
    for rank, (j, row) in enumerate(group.sort("amt_user_rating", ascending=False).iterrows()):
        if row.biped_num not in rankings:
            rankings.append(row.biped_num)
            all_biped_nums.add(row.biped_num)
    
    if len(rankings) > 0:
        biped_rankings.append(rankings)


global_rankings = list(all_biped_nums)
random.shuffle(global_rankings)
#global_rankings = [22, 16, 45, 39, 9, 6, 38, 48, 29, 1, 30, 23, 37, 8, 21, 44, 49, 10, 20, 13, 33, 4, 28, 19, 40, 47, 26, 41, 43, 25, 32, 11, 15, 17, 3, 36, 24, 5, 42, 18, 2, 14, 31, 12, 27, 34, 46, 35, 7]#list(all_biped_nums)

best_fitness = 0.0

for ranking in biped_rankings:
    best_fitness += kendalltau([x for x in global_rankings if x in ranking], ranking)[0]

repeats = 0

while repeats < 1000:
    cur_fitness = 0.0
    
    # try a new order
    new_global_rankings = list(global_rankings)
    (a, b) = np.random.randint(0, len(global_rankings), 2)
    #new_global_rankings[a], new_global_rankings[b] = new_global_rankings[b], new_global_rankings[a]
    move_val = new_global_rankings[a]
    del new_global_rankings[a]
    new_global_rankings.insert(b, move_val)
    
    for ranking in biped_rankings:
        cur_fitness += kendalltau([x for x in new_global_rankings if x in ranking], ranking)[0]
    
    if cur_fitness > best_fitness:
        repeats = 0
        best_fitness = cur_fitness
        global_rankings = new_global_rankings
        print best_fitness, "|", global_rankings
    else:
        repeats += 1
```

# Is a biped's rating influenced by the other bipeds it's presented with?

Are good bipeds rated better when shown with bad bipeds? Are bad bipeds rated worse when shown with good bipeds?

Look at each biped individually. Plot its rating from a rating set vs. the mean of the mean rating for the other bipeds in that set.

It looks like the background of bipeds doesn't affect how most bipeds are rated, except for bipeds 33 and 37.

Both of these bipeds appearing better in a background of bad bipeds makes sense: They both look somewhat promising (but not really).

```python
from scipy.stats import spearmanr

biped_mean_ratings = biped_ratings.groupby("biped_num")["amt_user_rating"].mean()

for biped_num in biped_mean_ratings.keys():
    
    amt_ratings = biped_ratings[biped_ratings["biped_num"] == biped_num]["amt_user_rating"].values
    background_ratings = []
    amt_rating_sets = biped_ratings[biped_ratings["biped_num"] == biped_num]["amt_rating_set"].values
    
    
    for rating_set in amt_rating_sets:
        set_mean_ratings = []
        for other_biped in biped_ratings[biped_ratings["amt_rating_set"] == rating_set]["biped_num"]:
            if other_biped == 0 or other_biped == biped_num:
                continue
                
            set_mean_ratings.append(biped_mean_ratings[other_biped])
            
        background_ratings.append(mean(set_mean_ratings))
    
    r, pval = spearmanr(background_ratings, amt_ratings)
    
    figure()
    title("Biped %d | r = %f, p = %f" % (biped_num, r, pval))
    scatter(x=background_ratings, y=amt_ratings)
    xlabel("Background biped ratings")
    ylabel("Biped rating")
```
