```python
from pandas import read_csv

biped_ratings = read_csv("biped-amt-ratings-training.csv", na_values=[""]).dropna()

# only keep the rating if the user correctly identified the dummy biped
biped_ratings = biped_ratings[biped_ratings["amt_user_red"] == biped_ratings["actual_red"]]

biped_ratings.describe()
```

# Which bipeds are highly rated?

Top 5 rated bipeds: 6, 16, 22, 37, 39

All of those except 6 result in good runs.

```python
data_lists = []
x_ticks = []

for i, group in biped_ratings.groupby("biped_num"):
    x_ticks.append(i)
    data_lists.append(group["amt_user_rating"].values)
    
figure(figsize=(15, 7))
boxplot(data_lists, notch=True)
xlabel("Biped number", fontsize=16)
ylabel("AMT user rating", fontsize=16)
xticks(range(1, 18), x_ticks);
```

# Is a biped's rating influenced by the other bipeds it's presented with?

Are good bipeds rated better when shown with bad bipeds? Are bad bipeds rated worse when shown with good bipeds?

Look at each biped individually. Plot its rating from a rating set vs. the mean of the mean rating for the other bipeds in that set.

It looks like the background of bipeds doesn't affect how most bipeds are rated, except for bipeds 33 and 37.

Both of these bipeds appearing better in a background of bad bipeds makes sense: They both look somewhat promising (but not really).

```python
from scipy.stats import spearmanr

biped_mean_ratings = biped_ratings.groupby("biped_num")["amt_user_rating"].mean()
del biped_mean_ratings[0]

for biped_num in biped_mean_ratings.keys():
    
    amt_ratings = biped_ratings[biped_ratings["biped_num"] == biped_num]["amt_user_rating"].values
    background_ratings = []
    amt_rating_sets = biped_ratings[biped_ratings["biped_num"] == biped_num]["amt_rating_set"].values
    
    
    for rating_set in amt_rating_sets:
        set_mean_ratings = []
        for other_biped in biped_ratings[biped_ratings["amt_rating_set"] == rating_set]["biped_num"]:
            if other_biped == 0 or other_biped == biped_num:
                continue
                
            set_mean_ratings.append(biped_mean_ratings[other_biped])
            
        background_ratings.append(mean(set_mean_ratings))
    
    r, pval = spearmanr(background_ratings, amt_ratings)
    
    figure()
    title("Biped %d | r = %f, p = %f" % (biped_num, r, pval))
    scatter(x=background_ratings, y=amt_ratings)
    xlabel("Background biped ratings")
    ylabel("Biped rating")
```

# Are ratings affected by AMT user specific variables?

AMT user age, gender, ID, nor time to complete the task seem to correlate with specific ratings.

```python
from scipy.stats import spearmanr

r, pval = spearmanr(biped_ratings["amt_user_age"], biped_ratings["amt_user_rating"])

figure(figsize=(15, 7))
title("r = %f, p = %f" % (r, pval))
scatter(x=biped_ratings["amt_user_age"], y=biped_ratings["amt_user_rating"])
xlabel("Age")
ylabel("Rating")
```

```python
biped_ratings.groupby("amt_user_gender")["amt_user_rating"].mean().plot(kind="barh")
```

```python
biped_ratings.groupby("amt_user_id")["amt_user_rating"].mean().plot(figsize=(15, 7))
```

```python
from scipy.stats import spearmanr

r, pval = spearmanr(biped_ratings["amt_user_time"], biped_ratings["amt_user_rating"])

figure(figsize=(15, 7))
title("r = %f, p = %f" % (r, pval))
scatter(x=biped_ratings["amt_user_time"], y=biped_ratings["amt_user_rating"])
xlabel("Time to complete task")
ylabel("Rating")
```
