# Load data

```python
%pylab inline
```

```python
from pandas import read_csv

biped_ratings = read_csv("biped-amt-ratings-training.csv", na_values=[""]).dropna()

# only keep the rating if the user correctly identified the dummy biped
biped_ratings = biped_ratings[biped_ratings["amt_user_red"] == biped_ratings["actual_red"]]

# throw out ratings for the dummy biped (biped 0)
biped_ratings = biped_ratings[biped_ratings["biped_num"] != 0]

biped_ratings.describe()
```

# Which bipeds are highly rated?

Top 5 rated bipeds: 6, 16, 22, 37, 39

All of those except 6 result in good runs.

## Distributions

```python
data_lists = []
x_ticks = []

for i, group in biped_ratings.groupby("biped_num"):
    x_ticks.append(i)
    data_lists.append(group["amt_user_rating"].values)
    
figure(figsize=(15, 7))
boxplot(data_lists, notch=True)
xlabel("Biped number", fontsize=16)
ylabel("AMT user rating", fontsize=16)
xticks(range(1, 18), x_ticks);
```

## 95% confidence intervals

```python
import scikits.bootstrap as bootstrap
import scipy

biped_nums = []

figure(figsize=(15, 7))

for i, group in biped_ratings.groupby("biped_num"):
    biped_nums.append(i)
    rating_mean = group["amt_user_rating"].mean()
    conf_int = bootstrap.ci(data=group["amt_user_rating"], statfunction=scipy.mean)
    errorbar(x=[i], y=[rating_mean], yerr=[[rating_mean - conf_int[0]], [conf_int[1] - rating_mean]], fmt="o")
    
xticks(biped_nums, fontsize=14)
yticks(fontsize=14)
xlabel("Biped number", fontsize=16)
ylabel("Combined AMT user rating", fontsize=16)
xlim(1, 48);
```

## Only lower bound of 95% CI

```python
import scikits.bootstrap as bootstrap
import scipy

biped_lower_bounds = {}

for i, group in biped_ratings.groupby("biped_num"):
    conf_int = bootstrap.ci(data=group["amt_user_rating"], statfunction=scipy.mean)
    biped_lower_bounds[i] = conf_int[0]
 
biped_nums = []
lower_bounds = []

for lb in sorted(biped_lower_bounds, key=biped_lower_bounds.get, reverse=True):
    biped_nums.append(lb)
    lower_bounds.append(biped_lower_bounds[lb])
 
figure(figsize=(15, 7))
plot(range(len(biped_nums)), lower_bounds, "o")
xticks(range(len(biped_nums)), biped_nums, fontsize=14)
yticks(fontsize=14)
xlabel("Biped number", fontsize=16)
ylabel("Lower 95% CI bound from AMT user ratings", fontsize=16)
xlim(-1, 17);
```

# Use rankings instead of specific numerical ratings

Infer rankings from specific numerical values. Each set has 5 ratings. Use those ratings to instead construct rankings.

More-or-less works out the same as going purely based on averages of the numerical ratings.

```python
from collections import defaultdict

biped_rankings = defaultdict(list)

for i, group in biped_ratings.groupby("amt_rating_set"):
    for rank, (j, row) in enumerate(group.sort("amt_user_rating", ascending=False)[group["biped_num"] != 0].iterrows()):
        biped_rankings[row["biped_num"]].append(rank + 1)
        
for biped in biped_rankings.keys():
    biped_rankings[biped] = mean(biped_rankings[biped])
```

```python
biped_nums = []
avg_rankings = []

for biped in sorted(biped_rankings, key=biped_rankings.get):
    biped_nums.append(biped)
    avg_rankings.append(biped_rankings[biped])
    
figure(figsize=(15, 7))
plot(range(len(biped_nums)), avg_rankings, "o")
xticks(range(len(biped_nums)), biped_nums, fontsize=14)
yticks(fontsize=14)
xlabel("Biped number", fontsize=16)
ylabel("Average ranking", fontsize=16)
xlim(-1, 17);
```

# Is a biped's rating influenced by the other bipeds it's presented with?

Are good bipeds rated better when shown with bad bipeds? Are bad bipeds rated worse when shown with good bipeds?

Look at each biped individually. Plot its rating from a rating set vs. the mean of the mean rating for the other bipeds in that set.

It looks like the background of bipeds doesn't affect how most bipeds are rated, except for bipeds 33 and 37.

Both of these bipeds appearing better in a background of bad bipeds makes sense: They both look somewhat promising (but not really).

```python
from scipy.stats import spearmanr

biped_mean_ratings = biped_ratings.groupby("biped_num")["amt_user_rating"].mean()

for biped_num in biped_mean_ratings.keys():
    
    amt_ratings = biped_ratings[biped_ratings["biped_num"] == biped_num]["amt_user_rating"].values
    background_ratings = []
    amt_rating_sets = biped_ratings[biped_ratings["biped_num"] == biped_num]["amt_rating_set"].values
    
    
    for rating_set in amt_rating_sets:
        set_mean_ratings = []
        for other_biped in biped_ratings[biped_ratings["amt_rating_set"] == rating_set]["biped_num"]:
            if other_biped == 0 or other_biped == biped_num:
                continue
                
            set_mean_ratings.append(biped_mean_ratings[other_biped])
            
        background_ratings.append(mean(set_mean_ratings))
    
    r, pval = spearmanr(background_ratings, amt_ratings)
    
    figure()
    title("Biped %d | r = %f, p = %f" % (biped_num, r, pval))
    scatter(x=background_ratings, y=amt_ratings)
    xlabel("Background biped ratings")
    ylabel("Biped rating")
```

# Are ratings affected by AMT user specific variables?

AMT user age, gender, ID, nor time to complete the task seem to correlate with specific ratings.

```python
from scipy.stats import spearmanr

r, pval = spearmanr(biped_ratings["amt_user_age"], biped_ratings["amt_user_rating"])

figure(figsize=(15, 7))
title("r = %f, p = %f" % (r, pval))
scatter(x=biped_ratings["amt_user_age"], y=biped_ratings["amt_user_rating"])
xlabel("Age")
ylabel("Rating")
```
