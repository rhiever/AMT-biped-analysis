# What predicts final fitness better: Early fitness or exploration metric?

```python
def compare_fitnesses(evaluation, domain, metric):
    data_subset = all_data[all_data.Evaluation == evaluation]
    data_subset = data_subset[data_subset.Domain == domain]
    data_subset_fitness = data_subset[data_subset.Selection == "Fitness"]
    data_subset_fitnov = data_subset[data_subset.Selection == "Fitnov"]
    data_subset_nov = data_subset[data_subset.Selection == "Novelty"]

    sb.jointplot(metric, "Final_Fitness", data=data_subset_fitness, kind="reg", color=sb.color_palette()[0])
    title(domain + " - Fitness")

    sb.jointplot(metric, "Final_Fitness", data=data_subset_fitnov, kind="reg", color=sb.color_palette()[1])
    title(domain + " - Fitnov")

    sb.jointplot(metric, "Final_Fitness", data=data_subset_nov, kind="reg", color=sb.color_palette()[2])
    title(domain + " - Novelty")
```

# Data wrangling

Columns:

* Evaluations so far (i.e. popsize x generation)

* best_fitness_so_far

* N/A

* exploration_metric

```python
!head randy-results/biped_fitness_100_log.txt
```

```python
from glob import glob
from pandas import read_csv, concat

experiment_dfs = []
for i, filename in enumerate(sorted(glob("randy-results/biped_fitness_*.txt"))):
    df = read_csv(filename, sep=" ", names=["Evaluation", "Cur_Best_Fitness", "NA", "Exploration_Metric", "Clock_Time"])
    del df["NA"]
    df["Run_Number"] = i + 1
    df["Domain"] = "Biped"
    df["Selection"] = "Fitness"
    df["Final_Fitness"] = max(df.Cur_Best_Fitness)
    experiment_dfs.append(df)
    
for i, filename in enumerate(sorted(glob("randy-results/biped_fitnov_*.txt"))):
    df = read_csv(filename, sep=" ", names=["Evaluation", "Cur_Best_Fitness", "NA", "Exploration_Metric", "Clock_Time"])
    del df["NA"]
    df["Run_Number"] = i + 1
    df["Domain"] = "Biped"
    df["Selection"] = "Fitnov"
    df["Final_Fitness"] = max(df.Cur_Best_Fitness)
    experiment_dfs.append(df)
    
for i, filename in enumerate(sorted(glob("randy-results/biped_novelty_*.txt"))):
    df = read_csv(filename, sep=" ", names=["Evaluation", "Cur_Best_Fitness", "NA", "Exploration_Metric", "Clock_Time"])
    del df["NA"]
    df["Run_Number"] = i + 1
    df["Domain"] = "Biped"
    df["Selection"] = "Novelty"
    df["Final_Fitness"] = max(df.Cur_Best_Fitness)
    experiment_dfs.append(df)

    
for i, filename in enumerate(sorted(glob("randy-results/maze_fitness_*.txt"))):
    df = read_csv(filename, sep=" ", names=["Evaluation", "Cur_Best_Fitness", "NA", "Exploration_Metric", "Clock_Time"])
    del df["NA"]
    df["Run_Number"] = i + 1
    df["Domain"] = "Maze"
    df["Selection"] = "Fitness"
    df["Final_Fitness"] = max(df.Cur_Best_Fitness)
    experiment_dfs.append(df)
    
for i, filename in enumerate(sorted(glob("randy-results/maze_fitnov_*.txt"))):
    df = read_csv(filename, sep=" ", names=["Evaluation", "Cur_Best_Fitness", "NA", "Exploration_Metric", "Clock_Time"])
    del df["NA"]
    df["Run_Number"] = i + 1
    df["Domain"] = "Maze"
    df["Selection"] = "Fitnov"
    df["Final_Fitness"] = max(df.Cur_Best_Fitness)
    experiment_dfs.append(df)
    
for i, filename in enumerate(sorted(glob("randy-results/maze_novelty_*.txt"))):
    df = read_csv(filename, sep=" ", names=["Evaluation", "Cur_Best_Fitness", "NA", "Exploration_Metric", "Clock_Time"])
    del df["NA"]
    df["Run_Number"] = i + 1
    df["Domain"] = "Maze"
    df["Selection"] = "Novelty"
    df["Final_Fitness"] = max(df.Cur_Best_Fitness)
    experiment_dfs.append(df)
    
all_experiment_df = concat(experiment_dfs)
all_experiment_df.to_csv("randy-results/all_experiments.csv", index=False)
```
