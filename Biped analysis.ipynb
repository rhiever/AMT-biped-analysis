# Can early generation fitness predict the final generation fitness?

The first generation that early fitness is significantly predictive of final fitness is generation ~150.

We therefore need input from AMT users before generation 150.

We should look at the videos of the early bipeds from the best runs at around generation 50 or 100, and compare them to the bad runs, to see if a human would actually be able to tell which one is more promising.

```python
from glob import glob
from pandas import read_csv
data_dfs = []

for fitness_file in sorted(glob("*_log.txt")):
    fitness_data = read_csv(fitness_file, sep=" ", names=["generation", "fitness", "dummy"])
    del fitness_data["dummy"]
    data_dfs.append(fitness_data)
```

```python
from scipy.stats import spearmanr

for early_gen in range(0, 1001, 50):
    early_fitnesses = []
    late_fitnesses = []
    for data_df in data_dfs:
        early_fitnesses.append(data_df.fitness[early_gen])
        late_fitnesses.append(data_df.fitness[2000])
        
    (corr_coeff, pval) = spearmanr(early_fitnesses, late_fitnesses)
    
    figure(figsize=(12, 7))
    title("gen %d vs. 2000, corr coeff = %f, p = %f" % (early_gen, corr_coeff, pval))
    scatter(early_fitnesses, late_fitnesses)
    xlabel("fitness at gen %d" % (early_gen))
    ylabel("fitness at gen 2000")
```
